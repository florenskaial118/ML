#Коллекции
**Коллекция** — структура данных (“переменная-контейнер”), которая включает в себя некоторое количество объектов `одного` или `разных` типов и позволяет обращаться к ним.

Встроенные в Python коллекции:
- список
- кортеж
- множество
- словарь

**Не путать с модулем collections**

`collections` - это модуль, который предоставляет специализированные типы данных, на основе словарей, кортежей, множеств, списков, т.е. модуль, содержащий набор “объектов-оберток” для различных типов коллекций.

Согласно документации нет понятия коллекций.
Однако, есть такие понятия, как:
- последовательности (sequences)
- множества (sets)
- отображения (mappings)

##Методы словарей

- `dict.clear()` - очищает словарь.
- `dict.copy()` - возвращает копию словаря (т.е. новый объект).
- `dict.fromkeys(seq[, value]`) - создает словарь с ключами из seq и значением value (по умолчанию None).
- `dict.get(key[, default])` - возвращает значение ключа, если такой есть, в противном случае не бросает исключение, а возвращает default (по умолчанию None).
- `dict.items()` - возвращает пары (ключ, значение).
- `dict.keys()` - возвращает ключи в словаре.
- `dict.pop(key[, default])` - удаляет ключ и возвращает значение. Если ключа нет, возвращает default (по умолчанию бросает исключение).
- `dict.popitem()` - удаляет и возвращает пару (ключ, значение). Если словарь пуст, бросает исключение KeyError. Помните, что словари неупорядочены.
- `dict.setdefault(key[, default])` - возвращает значение ключа, если такой есть, в противном случае не бросает исключение, а создает ключ со значением default (по умолчанию None).
- `dict.update([other])` - обновляет словарь, добавляя пары (ключ, значение) из other (т.е. обновляет объект). Существующие ключи перезаписываются. Возвращает None (не новый словарь!).
- `dict.values()` - возвращает значения в словаре.

````python
d = {a: a ** 2 for a in range(7)}
print(d)
d_2 = d
d_2[1] = 2
print(d)
print(d_2)
````

````python
d = {a: a ** 2 for a in range(7)}
print(d)
d_2 = d.copy()
d_2[1] = 2
print(d)
print(d_2)
````

````python
d = {a: a ** 2 for a in range(7)}
d['1']
````

````python
d = {a: a ** 2 for a in range(7)}
d.get('1', -1)
````

````python
d = {name: len(name) for name in ['Marchellis', 'Tisha', 'Luna', 'Mila', 'Karamelka']}
print(d.items())
print(d.keys())
print(d.values())
print(type(d.items()))
````

````python
d = {name: len(name) for name in ['Marchellis', 'Tisha', 'Luna', 'Mila', 'Karamelka']}
print(d)
print(d.pop('Mila'))
print(d)
````

````python
d = {name: len(name) for name in ['Marchellis', 'Tisha', 'Luna', 'Karamelka']}
print(d)
#d['Mila'] = len('Mila')
print(d.update({'Mila': len('Mila')}))
print(d)
````

## Оператор `match ... case ...`
**Начиная с Python 3.10**.

Синтаксис `math/case`:

```
match subject:
    case <pattern_1>:
        <action_1>
    case <pattern_2>:
        <action_2>
    case <pattern_3>:
        <action_3>
    case _:
        <action_wildcard>
```

**Пример:**
Необходимо напечатать полное название дня недели для введенного числа от 1 до 7.


````python
a = int(input())
if a == 1: print('Monday')
if a == 2: print('Tuesday')
if a == 3: print('Wednesday')
if a == 4: print('Thursday')
if a == 5: print('Friday')
if a == 6: print('Saturday')
if a == 7: print('Sunday')
````

````python
week_days = {
    1: 'Monday',
    2: 'Tuesday',
    3: 'Wednesday',
    4: 'Thursday',
    5: 'Friday',
    6: 'Saturday',
    7: 'Sunday'
}

a = int(input())
print(week_days[a])
````

````python
a = int(input())
match a:
    case 1: print('Monday')
    case 2: print('Tuesday')
    case 3: print('Wednesday')
    case 4: print('Thursday')
    case 5: print('Friday')
    case 6: print('Saturday')
    case 7: print('Sunday')
    case _: print('Error!!!')
````
